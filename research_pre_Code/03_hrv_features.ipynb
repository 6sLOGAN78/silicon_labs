```python
import neurokit2 as nk
import numpy as np
print("NeuroKit2 version:", nk.__version__)
def extract_hrv_features(ecg_signal, sampling_rate=500):

    try:

        # Detect R peaks
        signals, info = nk.ecg_process(
            ecg_signal,
            sampling_rate=sampling_rate
        )

        rpeaks = info["ECG_R_Peaks"]

        # Need enough beats
        if len(rpeaks) < 10:
            return None

        # Time-domain HRV
        hrv_time = nk.hrv_time(
            rpeaks,
            sampling_rate=sampling_rate,
            show=False
        )

        # Frequency-domain HRV
        hrv_freq = nk.hrv_frequency(
            rpeaks,
            sampling_rate=sampling_rate,
            show=False
        )

        return {

            "SDRR":
                float(hrv_time["HRV_SDNN"].iloc[0]),

            "RMSSD":
                float(hrv_time["HRV_RMSSD"].iloc[0]),

            "pRR50":
                float(hrv_time["HRV_pNN50"].iloc[0]),

            "TINN":
                float(hrv_time["HRV_TINN"].iloc[0]),

            "LF_Power":
                float(hrv_freq["HRV_LF"].iloc[0]),

            "HF_Power":
                float(hrv_freq["HRV_HF"].iloc[0]),

            "LF_HF":
                float(hrv_freq["HRV_LFHF"].iloc[0])
        }

    except Exception:

        return {
            "SDRR": np.nan,
            "RMSSD": np.nan,
            "pRR50": np.nan,
            "TINN": np.nan,
            "LF_Power": np.nan,
            "HF_Power": np.nan,
            "LF_HF": np.nan
        }
```

```python
import vitaldb
arr = vitaldb.load_case(
    22,
    ["SNUADC/ECG_II"],
    interval=1/500
)

ecg = arr[:,0]

features = extract_hrv_features(
    ecg[:15*60*500],
    sampling_rate=500
)

print(features)
```

```python
import time

start = time.time()

extract_hrv_features(
    ecg[:15*60*500],
    sampling_rate=500
)

print(time.time() - start)
```
